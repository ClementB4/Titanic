## LES FEMMES ET LES ENFANTS D’ABORD
### 1. Faire un test d’hypothèse pour savoir si oui ou non, les enfants ont été privilégiés lors du naufrage. 

###### Nous allons tester l'indépendance entre les deux features (survived, enfant) en utilisant le test chi2
##### H0:  les enfants n'ont pas été privilégiés lors du naufrage.
###### Nous supposerons que le risque alpha=0.05

<p> Nous devons d'abord choisir les enfants dans la feature 'Age_categories' </p>

```r
# prendre just le category enfant
train$enfant<-ifelse(train$Age_categories=='Infant',1,0)
```

```r
# scatter plot le relation entre enfant et survived
plot(train$enfant, train$Survived, main = " LES ENFANTS D’ABORD ",
     xlab = "Enfant", ylab = "Survived",
     pch = 19, frame = FALSE)
abline(lm(train$Survived~ train$enfant, data =train), col = "blue")
```

```r
chisq.test(train$Survived,train$enfant,correct=FALSE)
```

*** Avec Chi² on a la p-value = 6.505e-06 < 0.05, donc H0 n'est pas acceptée et  les enfants ont été privilégiés lors du naufrage***

### 3. L’INFLUENCE DU PRIX DU BILLET SUR LA SURVIE DES PASSAGERS
####  Faire un test d’hypothèse pour savoir si oui ou non, le prix du billet a une influence sur la survie d’un passager. 

###### Nous allons tester l'indépendance entre les deux features (survived, Fare) en utilisant le test chi2
##### H0:  le prix du billet n'a pas une influence sur la survie d’un passager.
###### Nous supposerons que le risque alpha=0.05

<p> Nous devons transformer la colonne 'Fare' de continue en discret </p>

```r
# calculer le mean de Fare 
mean(train$Fare)
```

```r
# Si tarif> 32 donc c'est 1, sinon c'est 0
train$prixBillet<-ifelse(train$Fare>32,1,0)
```

```r
#str(train)
```

```r
plot(train$prixBillet, train$Survived, main = "L’INFLUENCE DU PRIX DU BILLET SUR LA SURVIE DES PASSAGERS",
     xlab = "Le prix du billet", ylab = "Survived",
     pch = 19, frame = FALSE)
abline(lm(train$Survived~ train$prixBillet, data =train), col = "blue")
```

```r
chisq.test(train$Survived,train$prixBillet,correct=FALSE)
```

***Avec Chi² on a la p-value = 7.411e-13 < 0.05, donc H0 n'est pas acceptée et le prix du billet a une influence sur la survie d’un passager.***
